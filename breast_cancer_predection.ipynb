```python
# splitting the data and feature scaling
x=data.iloc[:,2:32].values
y=data.iloc[:,1].values
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20,random_state=0)
```

```python
#feature scaling
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.fit_transform(x_test)
```

```python
x_train
```

```python
y_train
```

```python
#traning the data
#logistic regresission    ----> supervised
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
def my_model(x_train,y_train):
     log=LogisticRegression(random_state=0)
     log.fit(x_train,y_train)


#decession tree classifier -----> supervised 
     tree=DecisionTreeClassifier(random_state=0)
     tree.fit(x_train,y_train)

# randomforest
     forest=RandomForestClassifier(n_estimators=10,random_state=0,criterion='entropy')
     forest.fit(x_train,y_train)


     print('the accurracy of logistic regression',log.score(x_train,y_train))
     print('the accurracy of Decision Tree classifier',tree.score(x_train,y_train))
     print('the accurracy of random forest classifier',forest.score(x_train,y_train))
     return log,tree,forest
     
```

```python
# accuracy of the model
m=my_model(x_train,y_train)
```

```python
from sklearn.metrics import confusion_matrix 
c=confusion_matrix(y_test, m[0].predict(x_test))
tp=c[0][0]
tn=c[1][1]
fp=c[1][0]
fn=c[0][1]
print('confusion matrix \n',c)
print('accuracy',(tp+tn)/(tp+tn+fp+fn))
```

```python
from sklearn.metrics import accuracy_score,classification_report
for i in range(len(m)):
     print("model is",i)
     print(accuracy_score(y_test,m[i].predict(x_test)))
     print(classification_report(y_test,m[i].predict(x_test)))
```
